```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Sample Data (simulating an eCommerce sales dataset)
data = {
    'Order Date': ['2024-10-01', '2024-10-02', '2024-10-02', '2024-10-03', '2024-10-04', '2024-10-05', '2024-10-05', '2024-10-06'],
    'Product': ['Laptop', 'Smartphone', 'Laptop', 'Headphones', 'Laptop', 'Smartphone', 'Headphones', 'Smartwatch'],
    'Price': [1200, 800, 1200, 200, 1200, 800, 200, 300],
    'Quantity': [1, 2, 1, 3, 2, 1, 5, 1],
    'Category': ['Electronics', 'Electronics', 'Electronics', 'Accessories', 'Electronics', 'Electronics', 'Accessories', 'Wearables']
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Convert 'Order Date' to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Create a 'Sales' column
df['Sales'] = df['Price'] * df['Quantity']

# 1. Descriptive Analysis
total_sales = df['Sales'].sum()
aov = df['Sales'].mean()
total_quantity = df['Quantity'].sum()

print(f"Total Sales: ${total_sales:.2f}")
print(f"Average Order Value (AOV): ${aov:.2f}")
print(f"Total Quantity Sold: {total_quantity}")

# 2. Sales by Date Analysis (Daily Sales)
daily_sales = df.groupby('Order Date')['Sales'].sum()

# Plotting Daily Sales as a Line Plot
plt.figure(figsize=(12, 6))
plt.plot(daily_sales.index, daily_sales.values, color='b', marker='o', linestyle='-')
plt.title('Daily Sales Trend')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# 3. Top Selling Products
top_products = df.groupby('Product')['Sales'].sum().sort_values(ascending=False).head(10)

# Plotting Top Products as a Bar Diagram
plt.figure(figsize=(10, 6))
top_products.plot(kind='bar', color='skyblue')
plt.title('Top 10 Products by Sales')
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()

# 4. Monthly Sales Analysis
df['Month'] = df['Order Date'].dt.to_period('M')
monthly_sales = df.groupby('Month')['Sales'].sum()

# Plotting Monthly Sales as a Bar Diagram
plt.figure(figsize=(12, 6))
monthly_sales.plot(kind='bar', color='lightgreen')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.show()

# 5. Sales by Category
category_sales = df.groupby('Category')['Sales'].sum()

# Plotting Sales by Category as a Bar Diagram
plt.figure(figsize=(10, 6))
category_sales.plot(kind='bar', color='orange')
plt.title('Sales by Category')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()

# 6. Top Selling Products by Quantity Sold
top_quantity_products = df.groupby('Product')['Quantity'].sum().sort_values(ascending=False).head(10)

# Plotting Top Quantity Sold as a Bar Diagram
plt.figure(figsize=(10, 6))
top_quantity_products.plot(kind='bar', color='salmon')
plt.title('Top 10 Products by Quantity Sold')
plt.xlabel('Product')
plt.ylabel('Quantity Sold')
plt.xticks(rotation=45)
plt.show()
```
